## Project Design

* User roles - Customer, Admin  
    * **Admin**  
        * Admin has authority to activate/deactivate customer accounts, as well as view orders placed by customers.  
        * Admin adds coupon codes that can be redeemed by the customers to avail discounts on products.  
    * **Customer**  
        * Customer can search for books by its name, author, isbn or category  
        * Can only see books that have their status == "show".  
        * Can add/remove books to/from wishlist.  
        * Can add/remove books to/from cart  
        * Can manage cart - add/remove/modify quantity and so on.
        * Must be logged in to place order.  
        * Can add shipping/billing addresses, which can be reused later  
        * Able to apply discount coupons in payment page.   
* Featured books section - Sponsored/advertised books are displayed  
* New books - displays the latest books.  
* **Category**:  
    - catId  
    - catName  
    - catDescription  
    - position  
    - status  
    - image  
    - createdAt  
&nbsp;&nbsp;Category is to be sorted by position, and only those category that have their status == "show" are to be made visible to customers.  
* **Book**:  
    - BookID  
    - CatID  
    - Title  
    - ISBN  
    - Year  
    - Price  
    - Description  
    - Position  
    - Status  
    - Image  
   
  



### Initializing dict for dataset

```python
Category_dict = {
    'CatID': [],
    'CatName': [],
    'CatDescription': [],
    'Position': [],
    'Status': [],
    'Image': [],
    'CreatedAt': []
}
```

## Populating data

```python
def random_date():
    start = dt.strptime('1/1/1973', '%m/%d/%Y')
    end = dt.strptime('10/13/2021', '%m/%d/%Y')
    
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    date = start + timedelta(seconds=random_second)
    return date.strftime('%Y/%m/%d')
```

```python
_pos = 1
for cat in Categories:
    _id = 'cat' + (str(_pos) if len(str(_pos))>1 else '0'+str(_pos))
    Category_dict['CatID'].append(_id)
    Category_dict['CatName'].append(cat)
    Category_dict['CatDescription'].append(catDescriptions[cat])
    Category_dict['Position'].append(_pos)
    Category_dict['Status'].append('Available')
    Category_dict['Image'].append(catImages[cat])
    _date = random_date() 
    Category_dict['CreatedAt'].append(random_date())
    _pos += 1
```

## Creating Dataframe

```python
Category_df = pd.DataFrame(Category_dict)
```

## Augmenting Books

```python
def AugmCatID(row):
    _catID = Category_df[Category_df['CatName'] == row].CatID.to_string().split(' ')[-1]
    return _catID
def randomizedDate(row):
    return random_date()
_pos = 0
def positions(row):
    global _pos
    _pos+=1
    return _pos
def setStatus(row):
    return "Available"
def setDescription(row):
    return ""
```

```python
data['CatID'] = data['Category'].apply(AugmCatID)
data['Year'] = data['CatID'].apply(randomizedDate)
data['Position'] = data['CatID'].apply(positions)
data['Status'] = data['CatID'].apply(setStatus)
data['Description'] = data['CatID'].apply(setDescription)
```
