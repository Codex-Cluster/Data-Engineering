### Initializing dict for dataset

```python
Category_dict = {
    'CatID': [],
    'CatName': [],
    'CatDescription': [],
    'Position': [],
    'Status': [],
    'Image': []
}
SubCategory_dict = {
    'SubID': [],
    'SubName': [],
    'SubDescription': [],
    'Position': [],
    'Status': [],
    'Image': [],
    'CatID': []
}
```

## Populating data

```python
_pos = 1
for cat in Categories:
    _id = 'cat' + (str(_pos) if len(str(_pos))>1 else '0'+str(_pos))
    Category_dict['CatID'].append(_id)
    Category_dict['CatName'].append(cat)
    Category_dict['CatDescription'].append(catDescriptions[cat])
    Category_dict['Position'].append(_pos)
    Category_dict['Status'].append('Available')
    Category_dict['Image'].append(catImages[cat])
    _pos += 1
```

```python
_pos = 1
for SubCat in SubCategories:
    _id = 'subCat' + (str(_pos) if len(str(_pos))>1 else '0'+str(_pos))
    SubCategory_dict['SubID'].append(_id)
    SubCategory_dict['SubName'].append(SubCat)
    SubCategory_dict['SubDescription'].append(SubCatDescriptions[SubCat])
    SubCategory_dict['Position'].append(_pos)
    SubCategory_dict['Status'].append('Available')
    SubCategory_dict['Image'].append(SubCatImages[SubCat])
    _catID = ''
    for idx, value in enumerate(categorySubCategoryMapping.values()):
        if(SubCat in value):
            _catID = 'cat' + (str(idx+1) if len(str(idx+1))>1 else '0'+str(idx+1))
    SubCategory_dict['CatID'].append(_catID)
    _pos += 1
```

## Creating Dataframe

```python
Category_df = pd.DataFrame(Category_dict)
SubCategory_df = pd.DataFrame(SubCategory_dict)
```
